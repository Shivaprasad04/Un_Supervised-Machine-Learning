# Hierarchical clustering

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\shiva\Downloads\crime_data.csv")
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.preprocessing import StandardScaler

df1 = pd.read_csv(r"C:\Users\shiva\Downloads\crime_data.csv")

features = df1.iloc[:, 1:]  

scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)

linkage_matrix = linkage(X_scaled, method='ward')

dendrogram(linkage_matrix, labels=df1.iloc[:, 0].tolist(), orientation='top', distance_sort='descending')
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Crime Categories")
plt.ylabel("Distance")
```

# K Means Clustering

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
x,y = make_blobs(n_samples = 500,n_features = 2,centers = 3,random_state = 3)
plt.scatter(x[:,0],x[:,1])
plt.grid(True)
```

```python
k = 3
 
clusters = {}
np.random.seed(23)
 
for idx in range(k):
    center = 2*(2*np.random.random((x.shape[1],))-1)
    points = []
    cluster = {
        'center' : center,
        'points' : []
    }
     
    clusters[idx] = cluster
clusters
```

```python
plt.scatter(x[:,0],x[:,1])
for i in clusters:
    center = clusters[i]['center']
    plt.scatter(center[0],center[1],marker = '*',c='r')
plt.show()
```

# DB scan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df5 = pd.read_csv(r"C:\Users\shiva\Downloads\Mall_Customers.csv")
df5.head()
```

```python
df5 = df5.iloc[:,[3,4]].values
df5
```

```python
plt.scatter(df5[:,0],df5[:,1])
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=30, n_init=10)
    kmeans.fit(df5)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps = 5,min_samples = 5)
```

```python
labels = dbscan.fit_predict(df5)
np.unique(labels)
```

```python
plt.scatter(df5[labels == -1,0],df5[labels == -1,1],c= 'b')
plt.scatter(df5[labels == 0,0],df5[labels == 0,1],c= 'g')
plt.scatter(df5[labels == 1,0],df5[labels == 1,1],c= 'r')
plt.scatter(df5[labels == 2,0],df5[labels == 2,1],c= 'y')
plt.scatter(df5[labels == 3,0],df5[labels == 3,1],c= 'black')
plt.scatter(df5[labels == 4,0],df5[labels == 4,1],c= 'pink')
plt.scatter(df5[labels == 5,0],df5[labels == 5,1],c= 'brown')
plt.scatter(df5[labels == 6,0],df5[labels == 6,1],c= 'silver')

plt.xlabel('Annual income')
plt.ylabel('Spending Score')
```

# PCA 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
cancer
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
dfx = pd.DataFrame(cancer['data'],columns = cancer['feature_names'])
```

```python
dfx.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(dfx)
scaled_data = scaler.transform(dfx)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
scaled_data.shape
```

```python
x_pca.shape
```

```python
plt.scatter(x_pca[:,0],x_pca[:,1],c = cancer['target'])
plt.xlabel('First PCA')
plt.ylabel('Second PCA')
```
